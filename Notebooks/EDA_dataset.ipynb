## Data cleaning


Clean the data by handling outliers in income variables and preparing the data for further analysis. Winsorize the 'median_household_income' and 'per_capita_income' columns to handle outliers, then scale the numerical features using standardization.


```python
import numpy as np
from scipy.stats.mstats import winsorize

# Winsorize income variables
df_cleaned = df.copy()  # Create a copy to avoid modifying the original DataFrame
for col in ['median_household_income', 'per_capita_income']:
    df_cleaned[col] = winsorize(df_cleaned[col], limits=[0.05, 0.05]) # Winsorize top and bottom 5%

# Identify numerical features for scaling (excluding those already handled)
numerical_cols = [col for col in df_cleaned.columns if df_cleaned[col].dtype in [np.float64, np.int64] and col not in ['median_household_income', 'per_capita_income']]

# Standardize numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_cleaned[numerical_cols] = scaler.fit_transform(df_cleaned[numerical_cols])
```

## Data analysis

### Subtask:
Analyze the relationship between demographic features and voting patterns. Calculate the correlation between demographic features and voting patterns using Pearson correlation, perform t-tests to compare demographic characteristics between Democrat and Republican counties, and identify predictive features based on the results.


```python
import pandas as pd
from scipy.stats import ttest_ind

# Correlation Analysis
correlation_matrix = df_cleaned[['democrat', 'republican', 'median_household_income', 'per_capita_income', 'persons_total', 'persons_male', 'persons_female']].corr()
display(correlation_matrix)

# Statistical Tests (t-tests)
democrat_counties = df_cleaned[df_cleaned['democrat'] > df_cleaned['republican']]
republican_counties = df_cleaned[df_cleaned['republican'] > df_cleaned['democrat']]

# Perform t-tests for selected features
features_to_test = ['median_household_income', 'per_capita_income', 'persons_total', 'persons_male', 'persons_female']
results = {}
for feature in features_to_test:
    t_statistic, p_value = ttest_ind(democrat_counties[feature], republican_counties[feature])
    results[feature] = (t_statistic, p_value)

for feature, (t_statistic, p_value) in results.items():
    print(f"T-test for {feature}:")
    print(f"  t-statistic = {t_statistic:.4f}")
    print(f"  p-value = {p_value:.4f}")
    if p_value < 0.05:
        print(f"  There is a statistically significant difference in {feature} between Democrat and Republican counties.")
    else:
        print(f"  There is no statistically significant difference in {feature} between Democrat and Republican counties.")

# Identify Predictive Features
print("\nPredictive Features:")
for feature, (t_statistic, p_value) in results.items():
    if p_value < 0.05:
        if correlation_matrix.loc[feature, 'democrat'] > 0:
            print(f"Higher {feature} is significantly correlated with Democrat votes.")
        else:
            print(f"Lower {feature} is significantly correlated with Democrat votes.")
        if correlation_matrix.loc[feature, 'republican'] > 0:
            print(f"Higher {feature} is significantly correlated with Republican votes.")
        else:
            print(f"Lower {feature} is significantly correlated with Republican votes.")
```

Visualize the relationships between demographic features and voting patterns, focusing on the key findings from the data analysis.  Create visualizations for the geographic distribution of voting patterns.
```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Heatmap of Correlation Matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Demographic Features and Voting Patterns')
plt.show()

# 2. Visualize t-test results
features = list(results.keys())
t_statistics = [result[0] for result in results.values()]
p_values = [result[1] for result in results.values()]

plt.figure(figsize=(10, 6))
plt.bar(features, t_statistics, color='skyblue', label='t-statistic')
plt.plot(features, p_values, marker='o', linestyle='-', color='red', label='p-value')  # Plot p-values as a line
plt.xlabel('Features')
plt.ylabel('Value')
plt.title('T-test Results for Demographic Features')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# 3. Geographic Visualization (Box plots by state)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='state', y='democrat', data=df_cleaned)
plt.title('Democrat Votes by State')
plt.xticks(rotation=90)

plt.subplot(1, 2, 2)
sns.boxplot(x='state', y='republican', data=df_cleaned)
plt.title('Republican Votes by State')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()

# 4. Additional Visualizations (Scatter plots)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='median_household_income', y='democrat', data=df_cleaned, color='blue')
plt.title('Median Household Income vs. Democrat Votes')

plt.subplot(1, 2, 2)
sns.scatterplot(x='median_household_income', y='republican', data=df_cleaned, color='red')
plt.title('Median Household Income vs. Republican Votes')

plt.tight_layout()
plt.show()
```

## Data clustering

Perform K-means clustering on the preprocessed data to identify potential subgroups with distinct voting behaviors. Perform K-means clustering on the preprocessed data to identify potential subgroups with distinct voting behaviors, and then analyze the characteristics of each cluster.


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine optimal k using the elbow method
inertia = []
silhouette_avg = []
k_range = range(2, 11) # test k values from 2 to 10
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_cleaned[['median_household_income', 'per_capita_income', 'persons_total', 'persons_male', 'persons_female', 'democrat', 'republican']])
    inertia.append(kmeans.inertia_)
    silhouette_avg.append(silhouette_score(df_cleaned[['median_household_income', 'per_capita_income', 'persons_total', 'persons_male', 'persons_female', 'democrat', 'republican']], kmeans.labels_))

plt.figure(figsize=(10, 5))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(k_range, silhouette_avg, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal k')
plt.show()

# Based on the plots, choose optimal k (e.g., k=3). In real-world scenarios, you should carefully check the plots to determine the best value.
optimal_k = 3

# Perform K-means clustering with the optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_cleaned['cluster'] = kmeans.fit_predict(df_cleaned[['median_household_income', 'per_capita_income', 'persons_total', 'persons_male', 'persons_female', 'democrat', 'republican']])

# Analyze cluster characteristics
cluster_means = df_cleaned.groupby('cluster').agg({
    'median_household_income': 'mean',
    'per_capita_income': 'mean',
    'persons_total': 'mean',
    'persons_male': 'mean',
    'persons_female': 'mean',
    'democrat': 'mean',
    'republican': 'mean'
})

display(cluster_means)
```
